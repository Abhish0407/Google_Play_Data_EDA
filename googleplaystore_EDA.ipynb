```python
# function to deal with the M and k present in Size column and converting it into int type
def size_process(item):
  if str(item)[-1] == "M":
    res = float(str(item).replace("M",""))
    res = res * 1024
    return res
  elif str(item)[-1] == "":
    res = float(str(item).replace("k",""))
    return res
  else:
    return str(np.nan)
```

```python
size_process("19M")
```

```python
df_copy["Size"] = df_copy["Size"].apply(size_process)
```
